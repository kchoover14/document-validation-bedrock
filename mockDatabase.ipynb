```python
from faker import Faker
import json
import random

fake = Faker()

def generate_biometric_log(is_malicious=False):
    log = {
        "user_id": fake.uuid4(),
        "timestamp": fake.iso8601(),
        "device_id": f"SENSOR-{random.randint(100, 999)}",
        "location": fake.city(),
        "access_level": random.choice(["Admin", "Staff", "Guest"]),
        "heart_rate_bpm": random.randint(60, 100) if not is_malicious else random.randint(150, 200),
        "document_type": "Passport",
        "verification_score": round(random.uniform(0.85, 0.99), 2) if not is_malicious else 0.45
    }
    return log
```
